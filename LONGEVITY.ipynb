### *What is IDHM Longevity?*



The IDHM Longevity (IDHM-L) is one of the three components of the **Municipal Human Development Index (IDHM)**, which is a composite measure of indicators of three dimensions of human development: longevity, education and income. The longevity component is measured by life expectancy at birth, calculated by an indirect method from data from the Demographic Censuses of IBGE. This indicator shows the average number of years that people would live from birth, maintaining the same mortality patterns observed in the reference year. As a maximum longevity parameter, 85 years is used, while 25 years is considered the minimum parameter. Therefore, if the municipality in question has a life expectancy at birth of 70 years, its IDHM-L will be calculated as follows:

(70 - 25) / (85 - 25) => 45 / 60 => IDHM-L = 0.750.

Hence, the municipality's IDHM-L will be 0.750.

# ***Does the average temperature where we live impact longevity?***

+ In this topic we will make a relationship between longitude/latitude with longevity.

## Importing the Dataset and Libraries: 📖

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BRAZIL_CITIES_CLEAN.csv')
df
```

## **First steps**📑

### ***We will use the following columns to generate our conclusions***

+ CITY/STATES
+ LAT/LONG
+ IDHM_Longevidade

```python
df_general = df[["CITY","STATE","LAT", "LONG","IDHM_Longevidade"]]
df_general
```

```python
df_longevity = df_general.sort_values(by='IDHM_Longevidade', ascending=False)
df_longevity = df_longevity[["CITY", "STATE", "IDHM_Longevidade"]].reset_index(drop=True)
df_longevity
```

```python
df_lat_long = df_general[["CITY", "STATE", "LAT", "LONG"]].reset_index(drop=True)
df_lat_long
```

## *Graphics analysis*📊

```python
#Graphic of the IDHM Longevity by state.

df_longevity.groupby('STATE').mean().plot(kind='bar', figsize=(10, 5), color='blue')
plt.ylim(0.6, 1)
plt.title('IDHM Longevity by State')
plt.ylabel('Longevity')
plt.xlabel('State')
```

```python
plt.scatter(df_general['LONG'], df_general['LAT'], color='darkgreen', s=4)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geographic Distribution of Data')
plt.show()
```

*It is noticed that both latitude and longitude of Brazilian municipalities have negative values.*

*This is due to the reason that brazil is located below the equator and to the left of the meridian of Greenwich*

## **Correlations**📖

```python
corrrelation_lat = df_general['LAT'].corr(df_general['IDHM_Longevidade'])
corrrelation_long = df_general['LONG'].corr(df_general['IDHM_Longevidade'])

print('Correlation between latitude and longevity: ', corrrelation_lat)
print(f'Correlation between longitude and longevity: ', corrrelation_long)
```

```python
sns.regplot(data=df_general, x='LAT', y='IDHM_Longevidade', color='yellow', line_kws={'color': 'red'})
plt.ylim(0.5, 1)
plt.xlabel('Longevity')
plt.ylabel('Latitude')
plt.title('Correlation between latitude and longevity')
plt.show()
```

```python
sns.regplot(data=df_general, x='LONG', y='IDHM_Longevidade', color='purple', line_kws={'color': 'red'})
plt.ylim(0.5, 1)
plt.xlabel('Longevity')
plt.ylabel('Longitude')
plt.title('Correlation between longitude and longevity')
plt.show()
```

> It is noticed that latitude has a greater correlation with longevity than longitude and latitude defines the climate of the city more than longitude.

> Therefore, from this point on, our analysis will be focused solely and exclusively on latitude, no longer on longitude.

## *Defining latitude intervals*📍

*Given that latitude significantly impacts the average temperature of a location, let's compare the longevity indices of three sectors located in distinct regions from north to south. By doing so, we might verify the influence of average temperature on life expectancy.*

```python
# calculating the range of the latitude
max_lat = df_general["LAT"].max()
min_lat = df_general["LAT"].min()
range_lat = max_lat - min_lat
sector_lat = range_lat/3
sector_lat
```

```python
# Northest cities
north_cities_lat = max_lat - sector_lat
north_cities_lat
```

```python
# Southeast cities
southeast_cities_lat = min_lat + sector_lat
southeast_cities_lat
```

```python
# selecting the cities between max_lat and north_cities_lat
northest_cities = df_general.loc[(df_general["LAT"] <= max_lat) & (df_general["LAT"] > north_cities_lat)]
northest_cities
```

```python
# selecting the cities between north_cities_lat and southeast_cities_lat
between_cities = df_general.loc[(df_general["LAT"] <= north_cities_lat) & (df_general["LAT"] > southeast_cities_lat)]
between_cities
```

```python
# selecting the cities between southeast_cities_lat and min_lat
southeast_cities = df_general.loc[(df_general["LAT"] >= min_lat) & (df_general["LAT"] < southeast_cities_lat)]
southeast_cities
```

## **Conclusions**✔️

```python
# calculating the average of the IDHM_Longevidade of the northest cities
mean_longevity_north_sector = northest_cities["IDHM_Longevidade"].mean()
# calculating the average of the IDHM_Longevidade of the between cities
mean_longevity_between_sector = between_cities["IDHM_Longevidade"].mean()
# calculating the average of the IDHM_Longevidade of the southeast cities
mean_longevity_south_sector = southeast_cities["IDHM_Longevidade"].mean()
print(f"The average of the IDHM_Longevidade of the northest cities, between cities and southeast cities are, respectively: \n{mean_longevity_north_sector}, {mean_longevity_between_sector} and {mean_longevity_south_sector}")
```

```python
# line plotting mean IDHM_Longevidade by sector
sns.lineplot(x = ["Northest", "Between", "Southeast"], y = [mean_longevity_north_sector, mean_longevity_between_sector, mean_longevity_south_sector], color='green')
plt.ylim(0.7, 0.9)
plt.xlabel('Sector')
plt.ylabel('IDHM_Longevidade')
plt.title('Mean IDHM_Longevidade by Sector')
plt.show()
```

> The aforementioned plot clearly indicates a noteworthy trend in life expectancy as we move towards the southern regions. This trend could potentially be attributed to the average temperature, but it is important to consider other influential factors such as socio-economic conditions, access to healthcare, and the quality of healthcare services, among various other factors.
